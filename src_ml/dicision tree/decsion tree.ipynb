```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris['DESCR']) #priting the inof of the dataset
```

```python
iris['target']
```

```python
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
X
```

```python
y=iris['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_classifier=DecisionTreeClassifier()
```

```python
tree_classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,7))
tree.plot_tree(tree_classifier,filled=True)
```

```python
y_pred=tree_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = print(confusion_matrix(y_test,y_pred))
cr= print(classification_report(y_test,y_pred))
```

```python
#Decision tree with prepruning 
```

```python
params = {    
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']}

```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
gird= GridSearchCV(treemodel,param_grid=params,cv=5,scoring='accuracy')
```

```python
gird.fit(X_train,y_train)
```

```python
gird.best_estimator_
```

```python
gird.best_score_
```

```python
y_pred=gird.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

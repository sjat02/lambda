```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
## Feature enecoding (Label encoding and Onehot encoding)

df.columns
```

```python
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
```

```python
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=12,test_size=0.25)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] =  le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] =  le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
#OneHotEncoding  -- Col transformer

from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],
                                     remainder='passthrough'
                                     )
```

```python
ct
```

```python
import sys
import numpy as np
X_train=ct.fit_transform(X_train)
X_test=ct.fit_transform(X_test)
```

```python
X_train.shape
```

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

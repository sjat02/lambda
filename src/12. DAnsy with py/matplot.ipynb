```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,4,5,6]

#create a line plot

plt.plot(x,y,color='red',linestyle='-.',marker='o',linewidth=0.8,markersize=6)
plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Speed vs Time')
plt.grid(True)
```

```python
#multiple plots
##sample data

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9,3))

plt.subplot(1,3,1)
plt.plot(x,y1,color='gray')
plt.title('plot1')

plt.subplot(1,3,2)
plt.plot(y1,x,color='gray')
plt.title('plot2')

plt.subplot(1,3,3)
plt.plot(y1,y2,color='gray')
plt.title('plot3')
```

```python
#barplot

categories=['A','B','C','D','E']
values=[5,6,10,6,6]

plt.bar(categories,values,width=0.4,color='black')
```

```python
#histogram

data=[1,2,3,4,5,6,7,8,4,5,6,6,7,8,1,1,1,1,1,1,1]

plt.hist(data,bins=5,edgecolor='black',color='orange')
```

```python
#scatter plot

x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,marker='*')
```

```python
# pie chart

labels=['A','B','C']
sizes=[30,20,10]
colors=['golde','yellow','black']
explod=(0.2,0.1,0.4)

plt.pie(sizes,labels=labels,explode=explod,shadow=True)
```

```python
import pandas as pd
sales_data_df=pd.read_csv('sales_data.csv')
sales_data_df.head()
```

```python
##plot total sales by products

sales_data_df.info()
```

```python
total_sales_by_product=sales_data_df.groupby('Product Category')['Total Revenue'].sum()
```

```python
total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
#plot sales trend

sales_trend=sales_data_df.groupby('Date')['Total Revenue'].sum()
```

```python
sales_trend=sales_trend.reset_index()
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```

```python
sales_trend.plot()
```

```python
import seaborn as sns
```

```python
#basic plotting with seaborn

tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lineplot(x='day',y='total_bill',data=tips)
```

```python
#categorical plots

sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='sex',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=5,kde=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
import numpy as np
data = np.random.normal(70, 10, 100)  # Random test scores

# KDE plot
sns.histplot(data, kde=True)  # Shade=True fills the area under the curve
plt.title("KDE Plot of Test Scores")
plt.xlabel("Score")
plt.ylabel("Density")
plt.show()
```
